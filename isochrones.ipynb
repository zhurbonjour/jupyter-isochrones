```python
import folium
import requests
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.environ.get("API_KEY")
```

Point - это координаты точки, от которой будет производиться расчет изохрон. Скопируйте и вставьте координаты из яндекс карт.

Time - это время, изохроны для которого необходимо получить. Time должно быть массивом, даже если необходимо получить изохроны для одного значения времени.

```python
# координаты точки из яндекс карт 
point = [59.838589, 30.093051]
# время в минутах
time = 5, 15, 25

rpoint = list(reversed(point))
time = list(map(lambda x: x * 60, time))
```

```python
# Определение карты и отметка на ней точки отсчета для изохрон
m = folium.Map(point, zoom_start=15)
marker = folium.Marker(point).add_to(m)
```

Далее установка нескольких желаемых профилей движения - они будут на карте изображены отдельными слоями. Ненужное закомментировать.

```python
profiles = [
    "driving-car",
    # "driving-hgv",
    "foot-walking",
    # "foot-hiking",
    "cycling-regular",
    # "cycling-road",
    # "cycling-mountain",
    # "cycling-electric",
    # "wheelchair"
]
```

По умолчанию все слои на карте выключены. Чтобы выбрать необходимый слой, наведите на иконку слоев в правом верхнем углу. Каждый слой отражает изохроны для своего профиля движения.

```python
body = {"locations":[rpoint],"range":time}
for profile in profiles:
    url = f"https://api.openrouteservice.org/v2/isochrones/{profile}"
    header = {"Authorization": API_KEY}
    isochrones = requests.post(url=url, json=body, headers=header).json()
    layer = folium.FeatureGroup(name=profile, show=False).add_to(m)
    for feature in isochrones["features"]:
        folium.GeoJson(feature).add_to(layer)
folium.LayerControl().add_to(m)
m
```
